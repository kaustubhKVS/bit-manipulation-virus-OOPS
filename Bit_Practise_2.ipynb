# FILE 1

```python
import re

def readFileToCodeBuffer():
    file1 = open("./IFile1.txt","r")
    content = file1.read()
    file1.close()
    list_content = re.split(r'( |/n/n)', content)
    clean_bits = ' '.join(list_content).split()
    return clean_bits
```

```python
test = readFileToCodeBuffer()
```

```python
test
```

```python
len(test)
```

```python
def countTotalBytes(file_data):
    list_using_comp = [var for var in file_data if len(str(var)) // 8 == 1]
    return int(len(list_using_comp))

countTotalBytes(test)
```

```python
def countTotalBits(file_data):
    num_bits = 0
    for var in file_data:
        num_bits = num_bits + len(str(var))
    return num_bits

countTotalBits(test)
```

```python
def countTrueBits(file_data):
    true_bits = 0
    for var in file_data:
        true_bits = true_bits + var.count('1')
        #print('Value Count is:' + str(var.count('1')) +  '---'  + 'Number of True Bits:' + str(true_bits))
    return true_bits
        
countTrueBits(test)
```

```python
def countFalseBits(file_data):
    false_bits = 0
    for var in file_data:
        false_bits = false_bits + var.count('0')
        #print('Value Count is:' + str(var.count('1')) +  '---'  + 'Number of True Bits:' + str(true_bits))
    return false_bits
        
countFalseBits(test)
```

```python
def countFalseBytes(file_data):
    false_bytes = 0
    for var in file_data:
        if ( var.count('1') <= var.count('0') ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            false_bytes = false_bytes + 1
    return false_bytes

countFalseBytes(test)
```

```python
def countTrueBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var.count('1') > var.count('0') ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countTrueBytes(test)
```

```python
def countAllTrueBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var == '11111111' ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countAllTrueBytes(test)
```

```python
def countAllFalseBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var == '00000000' ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countAllFalseBytes(test)
```

```python
def countRecurrentTrueBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var == '10101010' ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countRecurrentTrueBytes(test)
```

```python
def countRecurrentFalseBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var == '01010101' ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countRecurrentFalseBytes(test)
```

```python
def countSymmetricBytes(file_data):
    true_bytes = 0
    for var in file_data:
        if ( var == '11110000' or var == '00001111' ):
#             print('Value of 1 and 0 is:' + str(var.count('1')) +  '-'  + str(var.count('0')))
#             print('Value is:' + str(var))
            true_bytes = true_bytes + 1
    return true_bytes

countSymmetricBytes(test)
```

```python
def checkForMalformedByte(file_data):
    list_of_index = [file_data.index(var) for var in file_data if not len(str(var)) // 8 == 1]
    return (len(list_of_index),list_of_index)

checkForMalformedByte(test)
```

```python
def checkTrueBit(bit_value):
    if ( bit_value == '1' ):
        return True

checkTrueBit(1)
```
